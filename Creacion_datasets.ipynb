```python
import numpy as np
```

Para la generación de los conjuntos de entrenamiento finales, se utilizan una serie de funciones provistas en el repositorio de SAITS, en el archivo *gene_UCI_electricity_dataset.py*. Este programa realiza un escalado del conjunto de datos y lo divide en secuencias de tamaño programable, al igual que la cantidad de *missing values* artificiales generados.

Este programa crea un archivo denominado datasets.h5 en el directorio facilitado, con este contenido:
- X_train
- X_val
- X_val_hat
- X_val_indicating_mask
- X_val_missing_mask
- X_test
- X_test_hat
- X_test_indicating_mask
- X_test_missing_mask

Para el entrenamiento del modelo solo se van a utilizar el conjunto de train y test para la inferencia.
