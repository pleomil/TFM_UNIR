## SAITS

Para ejecutar SAITS es necesario instalar la librería PyPOTS e importar los módulos necesarios.

Para ejecutar el modelo únicamente hay que crear un objeto de clase SAITS con los parámetros seleccionados y llamar a la función fit de la clase.

Para la inferencia cambia el conjunto de datos seleccionado.

```python
!pip install pypots
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from pypots.data import load_specific_dataset, mcar, masked_fill
from pypots.imputation import SAITS, BRITS
from pypots.utils.metrics import cal_mae, cal_mre, cal_mse
```

#### MCAR30

```python
data = np.load('/content/drive/MyDrive/ColabNotebooks/TFM/generated_data_saits/mcar30/train.npy')
```

```python
X = data
num_samples = data.shape[0]


saits = SAITS(n_steps=100, n_features=37, n_layers=2, d_model=256, d_inner=128, n_heads=4, d_k=64, d_v=64, dropout=0.1, epochs=10)
dataset = {"X": X}
saits.fit(dataset)
```

```python
X_test = np.load('/content/drive/MyDrive/ColabNotebooks/TFM/generated_data_saits/mcar30/test_x_hat.npy')
X_test_intact = np.load('/content/drive/MyDrive/ColabNotebooks/TFM/generated_data_saits/mcar30/test_x.npy')
X_indicating_mask = np.load('/content/drive/MyDrive/ColabNotebooks/TFM/generated_data_saits/mcar30/test_indicating_mask.npy')
x_test = {'X':X_test}
```

```python
imputation = saits.impute(x_test)
mae = cal_mae(imputation, X_test_intact, X_indicating_mask)
mre = cal_mre(imputation, X_test_intact, X_indicating_mask)
mse = cal_mse(imputation, X_test_intact, X_indicating_mask)

print('MAE {}, MRE {}, MSE {}'.format(mae,mre,mse))
```
